# Use a pipeline as a high-level helper

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")
```

# test the classifier

```python
pipe("I love NLP")
```

# Load tweets data about Engergy transtion  

```python
import pandas as pd

df = pd.read_csv("hf://datasets/DSCI511G1/COP26_Energy_Transition_Tweets/energy_transition_tweets_final.csv")
df.head()
```

# 16K rows and 14 columns!

```python
df.shape
```

# Take a sample of 100 tweets

```python
sample_df = df.sample(100)
```

# utility function that takes a tweet and return the label

```python
def classify_tweet(tweet):
    return pipe(tweet)[0]["label"]
```

# classify sample tweets

```python
sample_df['emotion'] = sample_df['tweetContent'].apply(classify_tweet)
```

# check results

```python
sample_df
```

```python
sample_df['emotion'].value_counts().to_frame()
```
