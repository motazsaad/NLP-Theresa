```python
# 🌍 Search tool using Google Search API
search_tool = SerperDevTool()
```

```python
from crewai import LLM
llm = LLM(model="ollama/qwen2.5:latest", base_url="http://localhost:11434")
```

```python
# 👨‍💻 Web Researcher Agent
web_researcher = Agent(
    role="Web Researcher",
    goal="Find official website for a given company",
    backstory="An expert in online research, skilled in finding and verifying company information.",
    verbose=True,
    tools=[search_tool],
    llm=llm  # Assign Qwen as the LLM
)
```

```python
# 🎯 List of company names
company_names = ["Tesla", "Volkswagen", "OpenAI", "SpaceX"]

# 📌 Create tasks dynamically for each company
tasks = []
for company in company_names:
    task = Task(
        description=(
            f"Search and retrieve the official website for the company '{company}'."
            " Ensure that only the official website is selected, ignoring social media pages or third-party sites."
            " Return the result in dictionary format as {'company_name': company, 'website': website_url}."
        ),
        expected_output="A dictionary with the company name and its official website URL.",
        tools=[search_tool],
        agent=web_researcher,
    )
    tasks.append(task)
```

```python
# 🚀 Crew Execution
crew = Crew(
    agents=[web_researcher],
    tasks=tasks,
    process=Process.sequential  # Runs tasks one by one
)
```

```python
!ollama ps
```

```python
# 🔥 Start the Crew and get results
crew_output = crew.kickoff()
```

```python
crew_output
```

```python
import json

def extract_company_websites(crew_output):
    results = []

    for task_output in crew_output.tasks_output:
        try:
            # Parse the raw output as JSON
            parsed_data = json.loads(task_output.raw)

            # Ensure both "company_name" and "website" keys exist
            if "company_name" in parsed_data and "website" in parsed_data:
                results.append({
                    "company_name": parsed_data["company_name"],
                    "website": parsed_data["website"]
                })
            else:
                print(f"Skipping invalid JSON output: {task_output.raw}")

        except json.JSONDecodeError:
            print(f"Skipping unstructured output: {task_output.raw}")

    return results

# Assuming `crew_output` contains the CrewOutput object you provided
company_websites = extract_company_websites(crew_output)

# Print formatted JSON output
formatted_json = json.dumps(company_websites, indent=4)
print("\n🔗 Extracted Company Websites:\n", formatted_json)
```
